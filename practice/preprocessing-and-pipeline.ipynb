## Beginner Level

#### Problem 1: Feature Scaling

Dataset: Iris flowers dataset

Task: Create a pipeline using sklearn.pipeline.Pipeline to preprocess the features of the iris flowers dataset. Apply feature scaling using sklearn.preprocessing.StandardScaler and train a classifier (e.g., Logistic Regression) to classify the iris flowers into different species.

#### Problem 2: One-Hot Encoding

Dataset: Housing prices dataset

Task: Build a pipeline using sklearn.pipeline.Pipeline to preprocess the features of the housing prices dataset. Perform one-hot encoding using sklearn.preprocessing.OneHotEncoder to encode categorical features and train a regression model (e.g., Linear Regression) to predict house prices.

#### Problem 3: Missing Value Imputation

Dataset: Customer churn dataset

Task: Construct a pipeline using sklearn.pipeline.Pipeline to preprocess the features of the customer churn dataset. Handle missing values using sklearn.preprocessing.SimpleImputer and train a classifier (e.g., Random Forest) to predict customer churn.

#### Problem 4: Label Encoding

Dataset: Titanic survival dataset

Task: Create a pipeline using sklearn.pipeline.Pipeline to preprocess the features of the Titanic survival dataset. Apply label encoding using sklearn.preprocessing.LabelEncoder to encode categorical features, and train a classifier (e.g., Random Forest) to predict passenger survival.

#### Problem 5: Standardization

Dataset: Diabetes dataset

Task: Build a pipeline using sklearn.pipeline.Pipeline to preprocess the features of the diabetes dataset. Apply standardization using sklearn.preprocessing.StandardScaler, and train a regression model (e.g., Ridge Regression) to predict diabetes progression.

## Advanced Level:

#### Problem 11: Robust Scaling and Feature Selection

Dataset: Credit card default dataset

Task: Construct an advanced pipeline using sklearn.pipeline.Pipeline to preprocess the features of the credit card default dataset. Perform feature scaling with robustness to outliers using sklearn.preprocessing.RobustScaler, apply feature selection using sklearn.feature_selection.SelectKBest, and train a classifier (e.g., Support Vector Machine) to predict credit card defaults.

#### Problem 12: Custom Transformation and Stacking Classifiers
